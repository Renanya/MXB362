```python
# Import Necessary Packages
import leafmap
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_json("https://a5c7zwf7e5.execute-api.ap-southeast-2.amazonaws.com/dev/offences?locationType=Lga&startDate=9-28-2025&locationName=Brisbane&endDate=09-30-2025&format=JSON")
df.head()
```

```python
absgdf= gpd.read_file("./Data/ABS_Data/MB_2016_QLD.dbf")
absgdf.head(100)
```

```python
absgdf['longitude'] = absgdf["geometry"].centroid.x
absgdf['latitude'] = absgdf["geometry"].centroid.y
absgdf['MB_CODE16'] = absgdf['MB_CODE16'].astype(int)
print(absgdf["longitude"])
print(absgdf["latitude"])
```

```python
absgdf.dtypes
```

```python
crimedf= pd.merge(df, absgdf, left_on="ABS Meshblock", right_on = "MB_CODE16", how = "inner") 
crimedf.tail()
```

```python
crimedf2 = crimedf[["Type","Date","Postcode","longitude","latitude"]]
crimedf2
```

```python
m = leafmap.Map(center=(-27.4705, 153), zoom=11)
m.add_points_from_xy(crimedf2, x="longitude", y="latitude")
m
```
