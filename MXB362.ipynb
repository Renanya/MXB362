```python
# Import Necessary Packages
import leafmap
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_json("https://a5c7zwf7e5.execute-api.ap-southeast-2.amazonaws.com/dev/offences?locationType=Lga&startDate=9-23-2025&locationName=Brisbane&endDate=09-30-2025&format=JSON")
df.head()
```

```python
absgdf= gpd.read_file("./Data/ABS_Data/MB_2016_QLD.dbf")
absgdf.head(100)
```

```python
absgdf['longitude'] = absgdf["geometry"].centroid.x
absgdf['latitude'] = absgdf["geometry"].centroid.y
absgdf['MB_CODE16'] = absgdf['MB_CODE16'].astype(int)
print(absgdf["longitude"])
print(absgdf["latitude"])
```

```python
absgdf.dtypes
```

```python
crimedf= pd.merge(df, absgdf, left_on="ABS Meshblock", right_on = "MB_CODE16", how = "inner") 

crimedf.tail()
```

```python
crimedf2 = crimedf[["Type","Date","Postcode","longitude","latitude"]]
crimedf2.dtypes
```

```python
Arson = crimedf2[crimedf2["Type"]=="Arson"].reset_index(drop=True)
Assault = crimedf2[crimedf2["Type"]=="Assault"].reset_index(drop=True)
Drugs = crimedf2[crimedf2["Type"]=="Drug Offences"].reset_index(drop=True)
Fraud = crimedf2[crimedf2["Type"]=="Fraud"].reset_index(drop=True)
Gaming = crimedf2[crimedf2["Type"]=="Gaming Racing & Betting Offences"].reset_index(drop=True)
GoodOrder = crimedf2[crimedf2["Type"]=="Good Order Offences"].reset_index(drop=True)
Stolen = crimedf2[crimedf2["Type"]=="Handling Stolen Goods"].reset_index(drop=True)
Homicide = crimedf2[crimedf2["Type"]=="Homicide (Murder)"].reset_index(drop=True)
OtherHomicide = crimedf2[crimedf2["Type"]=="Other Homicide"].reset_index(drop=True)
Liquor = crimedf2[crimedf2["Type"]=="Liquor (excl. Drunkenness)"].reset_index(drop=True)
Misc = crimedf2[crimedf2["Type"]=="Miscellaneous Offences"].reset_index(drop=True)
Other = crimedf2[crimedf2["Type"]=="Other Offences Against the Person"].reset_index(drop=True)
Property = crimedf2[crimedf2["Type"]=="Other Property Damage"].reset_index(drop=True)
OtherTheft = crimedf2[crimedf2["Type"]=="Other Theft (excl. Unlawful Entry)"].reset_index(drop=True)
Prostitution = crimedf2[crimedf2["Type"]=="Prostitution Offences"].reset_index(drop=True)
Robbery = crimedf2[crimedf2["Type"]=="Robbery"].reset_index(drop=True)
Stock = crimedf2[crimedf2["Type"]=="Stock Related Offences"].reset_index(drop=True)
Traffic = crimedf2[crimedf2["Type"]=="Traffic and Related Offences"].reset_index(drop=True)
Trespassing = crimedf2[crimedf2["Type"]=="Trespassing and Vagrancy"].reset_index(drop=True)
UnlawfulEntry = crimedf2[crimedf2["Type"]=="Unlawful Entry"].reset_index(drop=True)
Motor = crimedf2[crimedf2["Type"]=="Unlawful Use of Motor Vehicle"].reset_index(drop=True)
Weapons = crimedf2[crimedf2["Type"]=="Weapons Act Offences"].reset_index(drop=True)
```
